## Python requests, APIs, and JSON traversals to answer a fundamental question: 
# "What's the weather like as we approach the equator?"

Observations:

1) The hottest cities are currently around 85 degrees F. and are located around 20 degrees North.

2) In my original sample of cities, one city had an error in the humidity data that resulted in it being an outlier at 290% humidity.

3) Once outlier was excluded, the varience in humidity levels was consistent across latitudes.

4) Cloud cover had cosistent varience across latitudes.

5) Wind speeds had cosistent varience across latitudes.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys_2

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "/output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination using citipy
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

Your final notebook must:

Randomly select at least 500 unique (non-repeat) cities based on latitude and longitude.

Perform a weather check on each of the cities using a series of successive API calls.

Include a print log of each city as it's being processed with the city number and city name.

Save both a CSV of all data retrieved and png images for each scatter plot.

```python
# OpenWeatherMap API Key
api_key = api_keys_2.my_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?APPID=" + api_key 

# Save config information.
units = "imperial"
```

```python
# Build query URLs in loop
urls = []
for i in cities:
    city = i
    query_url = f"{url}&units={units}&q={city}"
    urls.append(query_url)
#     print(query_url)
```

```python
print(query_url)
```

```python
# Get weather data...
# store in lists
temp = []
humid = []
clouds = []
wind = []
city_failures = []
city_successes = []
lng_successes = []
lat_successes = []

# # single data call
# weather_response = requests.get(query_url)
# weather_json = weather_response.json()

# log legend
print('City Log')
print('T = Temperature in Farenheit; H = Humidity level; C = Cloudiness; W = Wind speed')
print('===============================================================================')

#call data in loop and log
for i in cities:
    city = i
    list_position = cities.index(i)+1
    query_url = f"{url}&units={units}&q={city}"
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()

    # store data
    try:
        temp.append(weather_json['main']['temp'])
        humid.append(weather_json['main']['humidity'])
        clouds.append(weather_json['clouds']['all'])
        wind.append(weather_json['wind']['speed'])
        city_successes.append(city)
        lng_successes.append(weather_json['coord']['lon'])
        lat_successes.append(weather_json['coord']['lat'])
        print(f"{list_position}. {city}, id:{weather_json['id']}, lat:{weather_json['coord']['lat']} lng:{weather_json['coord']['lon']};  weather: T= {weather_json['main']['temp']} F.; \
H= {weather_json['main']['humidity']}%; \
C= {weather_json['clouds']['all']}%; \
W= {weather_json['wind']['speed']} mph.")
    except:
        print(f"ERROR: weather call for {list_position}. {city} did not work.")
        city_failures.append(city)
    
# # print data for a city
# print(f"T: {weather_json['main']['temp']} deg. F;")
# print(f"H: {weather_json['main']['humidity']}%.")
# print(f"C: {weather_json['clouds']['all']}%.")
# print(f"W: {weather_json['wind']['speed']} mph.")
```

```python
# checking list lengths
print(len(temp))
print(len(humid))
print(len(clouds))
print(len(wind))
print(len(city_failures))
```

## Your objective is to build a series of scatter plots to showcase the following relationships:

Temperature (F) vs. Latitude

Humidity (%) vs. Latitude

Cloudiness (%) vs. Latitude

Wind Speed (mph) vs. Latitude

```python

# create a data frame from cities, lat, and temp
weather_dict = {
    "city": city_successes,
    "latitude": lat_successes,
    "longitutde": lng_successes,
    "temperature": temp,
    'humidity': humid,
    'cloudiness': clouds,
    'wind_speed': wind
}

weather_data = pd.DataFrame(weather_dict)
weather_data.to_csv("cities.csv")
```

```python
# Build a scatter plot for each data type
plt.scatter(weather_data["latitude"], weather_data["temperature"], marker="o")

# Incorporate the other graph properties
plt.title("Current Temperature in World Cities by Latitude")
plt.ylabel("Temperature (Farenheit)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("TemperatureInWorldCities.png")

# Show plot
plt.show()
```

```python
# Checking plot for outliers like in first sample of cities

# Build a scatter plot for each data type
plt.scatter(weather_data["latitude"], weather_data["humidity"], marker="o")

# Incorporate the other graph properties
plt.title("Current Humidity in World Cities by Latitude")
plt.ylabel("% Humidity")
plt.xlabel("Latitude")
plt.grid(True)

# Show plot
plt.show()
```

```python
# filtering out humidity outliers
cleaned_weather_df = weather_data.loc[weather_data['humidity']<=100 ,:]

# Build a scatter plot for each data type
plt.scatter(cleaned_weather_df["latitude"], cleaned_weather_df["humidity"], marker="o")

# Incorporate the other graph properties
plt.title("Current Humidity in World Cities by Latitude")
plt.ylabel("% Humidity")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("HumidityInWorldCities.png")

# Show plot
plt.show()
```

```python
# Build a scatter plot for each data type
plt.scatter(weather_data["latitude"], weather_data["cloudiness"], marker="o")

# Incorporate the other graph properties
plt.title("Current Cloudiness in World Cities by Latitude")
plt.ylabel("% Cloud Cover")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("CloudinessInWorldCities.png")

# Show plot
plt.show()
```

```python
# Build a scatter plot for each data type
plt.scatter(weather_data["latitude"], weather_data["wind_speed"], marker="o")

# Incorporate the other graph properties
plt.title("Current Wind Speed in World Cities by Latitude")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("WindSpeedInWorldCities.png")

# Show plot
plt.show()
```
